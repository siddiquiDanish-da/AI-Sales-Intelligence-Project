```python
# ============================================
# NOTEBOOK 5 — Generative AI Report
# Sales Intelligence Project
# Analyst: Danish Siddiqui
# ============================================

import pandas as pd
import json
import os

os.makedirs('ai_reports', exist_ok=True)

# Load clean data
sales = pd.read_csv('data/sales_clean.csv')
sales['date'] = pd.to_datetime(sales['date'])

# ── Build KPI Summary ─────────────────────────
total_revenue   = sales['revenue'].sum()
total_orders    = len(sales)
avg_order_value = sales['revenue'].mean()
total_units     = sales['quantity'].sum()

top_product = (sales.groupby('product_name')['revenue']
                   .sum().idxmax())
top_region  = (sales.groupby('region_name')['revenue']
                   .sum().idxmax())
best_month  = (sales.groupby('month_name')['revenue']
                   .sum().idxmax())

top3_products = (sales.groupby('product_name')['revenue']
                     .sum()
                     .sort_values(ascending=False)
                     .head(3))

top3_regions = (sales.groupby('region_name')['revenue']
                    .sum()
                    .sort_values(ascending=False)
                    .head(3))

low3_products = (sales.groupby('product_name')['revenue']
                     .sum()
                     .sort_values(ascending=True)
                     .head(3))

print("✅ KPI Summary ready!")
print(f"\n  Total Revenue   : ₹{total_revenue:,.0f}")
print(f"  Total Orders    : {total_orders}")
print(f"  Avg Order Value : ₹{avg_order_value:,.0f}")
print(f"  Top Product     : {top_product}")
print(f"  Top Region      : {top_region}")
print(f"  Best Month      : {best_month}")
```

```python
# ── Build Prompt for ChatGPT ──────────────────

top3_text = "\n".join([f"    {i+1}. {name}: ₹{rev:,.0f}"
                       for i, (name, rev)
                       in enumerate(top3_products.items())])

low3_text = "\n".join([f"    {i+1}. {name}: ₹{rev:,.0f}"
                       for i, (name, rev)
                       in enumerate(low3_products.items())])

top3_region_text = "\n".join([f"    {i+1}. {name}: ₹{rev:,.0f}"
                               for i, (name, rev)
                               in enumerate(top3_regions.items())])

prompt = f"""
You are a Senior Business Analyst at a top IT company.
Write a formal Monthly Sales Performance Report based on
the following KPI data. The report should be professional,
detailed, and include actionable recommendations.

=== SALES DATA SUMMARY ===
Total Revenue      : ₹{total_revenue:,.0f}
Total Orders       : {total_orders}
Average Order Value: ₹{avg_order_value:,.0f}
Total Units Sold   : {total_units:,}
Best Month         : {best_month}
Top Region         : {top_region}

Top 3 Products by Revenue:
{top3_text}

Underperforming Products:
{low3_text}

Top 3 Regions by Revenue:
{top3_region_text}

=== REPORT REQUIREMENTS ===
Write the report with these sections:
1. Executive Summary (3-4 sentences)
2. Revenue Performance Analysis
3. Product Performance Highlights
4. Regional Performance Analysis
5. Key Findings & Anomalies
6. Strategic Recommendations (at least 3)
7. Conclusion

Use professional business language.
Format with clear section headings.
Make it sound like a real company report.
"""

print("✅ AI Prompt built successfully!")
print("\nPrompt preview (first 300 chars):")
print(prompt[:300] + "...")

# Save prompt to file
with open('ai_reports/prompt.txt', 'w', encoding='utf-8') as f:
    f.write(prompt)
print("\n✅ Prompt saved → ai_reports/prompt.txt")
```

```python
# ── AI Generated Report — Saved ───────────────
import os
os.makedirs('ai_reports', exist_ok=True)
ai_report = """
Monthly Sales Performance Report

1. Executive Summary

The monthly sales performance demonstrates strong overall revenue generation, 
achieving a total of ₹134.23 million across 502 orders. The business recorded 
a healthy Average Order Value (AOV) of ₹267,383, indicating a focus on 
high-value transactions. December emerged as the best-performing month, with 
Delhi leading as the top revenue-generating region. While flagship products 
continue to drive growth, certain product categories show signs of 
underperformance, requiring strategic attention.

2. Revenue Performance Analysis

The organization achieved total revenue of ₹134,226,220, supported by 4,899 
units sold. The high AOV suggests a premium product mix and effective upselling 
strategies. Revenue concentration appears skewed toward a few high-performing 
products, which, while beneficial in the short term, may pose a diversification 
risk.

The strong performance in December indicates seasonal demand trends, likely 
driven by festive or year-end purchasing behavior. This presents an opportunity 
to further capitalize on seasonal campaigns in future cycles.

3. Product Performance Highlights

Top Performing Products:
- Laptop Pro leads with Rs 82.46 million in revenue
- Tablet X follows with Rs 36.35 million
- Smart Watch generated Rs 9.93 million

Underperforming Products:
- USB-C Hub: Rs 2.09 million
- Wireless Earbuds: Rs 3.39 million
- Smart Watch shows inconsistent performance across segments

4. Regional Performance Analysis

- Delhi leads with Rs 20.64 million in revenue
- Mumbai follows closely at Rs 20.35 million
- Kolkata contributes Rs 19.03 million

The narrow gap between top regions indicates balanced geographic distribution.

5. Key Findings and Anomalies

- Heavy revenue dependency on Laptop Pro raises concentration risk
- Smart Watch shows contradictory performance across regions
- Underperformance of accessory products suggests weak cross-selling
- Seasonal spike in December highlights reliance on limited high-demand periods
- Regional performance lacks significant differentiation

6. Strategic Recommendations

1. Diversify Product Revenue Streams - Reduce dependency on flagship products
   by promoting mid-tier and accessory products through bundling campaigns.

2. Optimize Underperforming Product Strategy - Reassess pricing and promotion
   strategies for USB-C Hub and Wireless Earbuds.

3. Leverage Seasonal Trends - Develop structured marketing campaigns around
   peak months like December with early promotions and inventory planning.

4. Enhance Regional Expansion - Explore growth in Tier-2 markets to expand
   customer base and reduce regional concentration.

5. Improve Smart Watch Sales Consistency - Conduct deeper analysis to identify
   performance gaps across regions and implement corrective measures.

7. Conclusion

Overall, the company demonstrates strong financial performance with high-value 
sales and consistent regional contributions. By focusing on diversification, 
targeted marketing, and regional expansion, the organization can achieve more 
sustainable and balanced growth in upcoming periods.

Generated by  : ChatGPT (Generative AI)
Analyst       : Danish Siddiqui
Project       : AI-Powered Sales Intelligence Platform
"""

# ── Save as Text File ─────────────────────────
with open('ai_reports/monthly_sales_report.txt',
          'w', encoding='utf-8') as f:
    f.write("=" * 60 + "\n")
    f.write("   MONTHLY SALES PERFORMANCE REPORT\n")
    f.write("   Generated by: Generative AI (ChatGPT)\n")
    f.write("   Analyst: Danish Siddiqui\n")
    f.write("   Project: AI-Powered Sales Intelligence Platform\n")
    f.write("=" * 60 + "\n\n")
    f.write(ai_report)

print("✅ AI Report saved → ai_reports/monthly_sales_report.txt")
print(f"   Total characters : {len(ai_report):,}")
print(f"   Total words      : {len(ai_report.split()):,}")
print("\n   Report contains:")
print("   ✅ Executive Summary")
print("   ✅ Revenue Performance Analysis")
print("   ✅ Product Performance Highlights")
print("   ✅ Regional Performance Analysis")
print("   ✅ Key Findings & Anomalies")
print("   ✅ Strategic Recommendations")
print("   ✅ Conclusion")
```
